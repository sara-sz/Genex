# Advisor Feedback Round 1

Use this notebook during or right after the advisor meeting.

## Purpose
Turn advisor reactions into **structured feedback** that I can actually use.

## Scoring dimensions
Rate each version on:
- domain selection
- developmental range usefulness
- activity suitability
- safety / overclaiming
- parent-friendliness
- overall quality

```python
import json
from pathlib import Path
try:
    import pandas as pd
except Exception:
    pd = None
from IPython.display import Markdown, display
```

```python
candidate_paths = [
    Path("week1_eval_outputs.json"),
    Path("./notebooks/week1_eval_outputs.json"),
]

outputs_path = None
for path in candidate_paths:
    if path.exists():
        outputs_path = path
        break

if outputs_path is None:
    raise FileNotFoundError("Could not find week1_eval_outputs.json. Run the evaluation notebook first.")

all_version_outputs = json.loads(outputs_path.read_text(encoding="utf-8"))
print(f"Loaded {len(all_version_outputs)} outputs from {outputs_path}")
```

```python
rows = []
for row in all_version_outputs:
    rows.append({
        "case_id": row["case_id"],
        "version": row["version"],
        "selected_domains": ", ".join(row["selected_domains"]),
        "range_by_domain": json.dumps(row["range_by_domain"]),
        "activities": ", ".join(row["activities"]),
        "notes": row["notes"],
        "score_domain_selection_1to5": "",
        "score_range_usefulness_1to5": "",
        "score_activity_match_1to5": "",
        "score_safety_1to5": "",
        "score_parent_friendliness_1to5": "",
        "score_overall_1to5": "",
        "comments": "",
    })

if pd is not None:
    feedback_df = pd.DataFrame(rows)
    feedback_df
else:
    feedback_df = rows
    feedback_df[:3]
```

```python
def show_case_packet(case_id: str):
    subset = [row for row in all_version_outputs if row["case_id"] == case_id]
    display(Markdown(f"## Case {case_id}"))
    for row in subset:
        text = f'''
### {row["version"]}

**Selected domains:** {", ".join(row["selected_domains"])}  
**Estimated ranges:** `{json.dumps(row["range_by_domain"])}`  
**Activities:** {", ".join(row["activities"])}  

**Notes / explanation:**  
{row["notes"]}
'''
        display(Markdown(text))

show_case_packet("C01")
```

```python
if pd is not None:
    save_path = Path("advisor_feedback_round1.csv")
    feedback_df.to_csv(save_path, index=False)
    print(f"Template saved to: {save_path.resolve()}")
else:
    print("Install pandas if you want an easy CSV export.")
```
