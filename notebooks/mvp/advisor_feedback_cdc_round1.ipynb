# Advisor Feedback Round 1 — CDC Brain

This notebook loads the outputs from `week1_cdc_brain.ipynb` and creates a ranking table for your advisor.

## What it gives you
- a summary table for all 5 children
- a detailed packet per child
- a blank CSV scoring sheet with one row per child

## Suggested scoring
Ask your advisor to rate each case on:
- estimated developmental ranges
- activity appropriateness
- safety / overclaiming
- usefulness
- overall quality

```python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Markdown, display
```

```python
def load_outputs():
    candidate_paths = [
        Path("week1_cdc_outputs.json"),
        Path("./notebooks/week1_cdc_outputs.json"),
        Path("../week1_cdc_outputs.json"),
        Path("../../week1_cdc_outputs.json"),
        Path("week1_cdc_outputs_sample.json"),
    ]
    for path in candidate_paths:
        if path.exists():
            return json.loads(path.read_text(encoding="utf-8")), path
    raise FileNotFoundError("Could not find week1_cdc_outputs.json. Run the CDC brain notebook first.")

results, results_path = load_outputs()
print("Loaded outputs from:", results_path.resolve())
print("Cases:", len(results))
```

```python
summary_rows = []
for result in results:
    summary_rows.append({
        "case_id": result["case_id"],
        "name": result["name"],
        "condition": result["condition"],
        "chronological_age_months": result["chronological_age_months"],
        "speech_range": result["developmental_profile"]["speech_language"]["estimated_range_months"],
        "motor_range": result["developmental_profile"]["motor"]["estimated_range_months"],
        "social_range": result["developmental_profile"]["social_communication"]["estimated_range_months"],
        "adaptive_range": result["developmental_profile"]["adaptive_cognitive"]["estimated_range_months"],
    })
summary_df = pd.DataFrame(summary_rows)
summary_df
```

```python
# Blank scoring table for the advisor.
feedback_rows = []
for result in results:
    feedback_rows.append({
        "case_id": result["case_id"],
        "name": result["name"],
        "condition": result["condition"],
        "score_ranges_1to5": "",
        "score_activities_1to5": "",
        "score_safety_1to5": "",
        "score_usefulness_1to5": "",
        "score_overall_1to5": "",
        "comments": "",
    })

feedback_df = pd.DataFrame(feedback_rows)
feedback_df
```

```python
def show_case_packet(case_id):
    matched = [r for r in results if r["case_id"] == case_id]
    if not matched:
        raise ValueError(f"No case found for {case_id}")
    result = matched[0]

    display(Markdown(f"# {result['case_id']} — {result['name']}"))
    display(Markdown(
        f"**Condition:** {result['condition']}  \
"
        f"**Chronological age (months):** {result['chronological_age_months']}  \
"
        f"**CDC start month:** {result['cdc_start_month']}  \
"
        f"**Concern:** {result['concern_text']}"
    ))

    for domain_key, profile in result["developmental_profile"].items():
        domain_title = domain_key.replace("_", " ").title()
        display(Markdown(f"## {domain_title}"))
        display(Markdown(
            f"**Estimated range:** {profile['estimated_range_months']} months  \
"
            f"**Confidence:** {profile['confidence']}"
        ))

        asked_lines = []
        for item in profile["asked"]:
            asked_lines.append(f"- [{item['band_month']}m] {item['question']} → **{item['answer']}**")
        display(Markdown("### Asked CDC questions\n" + "\n".join(asked_lines)))

        activity_lines = []
        for act in result["activities_by_domain"][domain_key]:
            activity_lines.append(
                f"**{act.get('activity_number')}. {act.get('title')}** ({act.get('duration_min')} min)  \
"
                f"Why: {act.get('why_this_fits')}  \
"
                f"Materials: {act.get('materials')}  \
"
                f"Instructions: {act.get('instructions')}  \
"
                f"Adaptation: {act.get('adaptation')}"
            )
        display(Markdown("### Suggested activities\n\n" + "\n\n".join(activity_lines)))

show_case_packet("C01")
```

```python
# Save the blank feedback table so your advisor can fill it out after the meeting.
save_path = Path("advisor_feedback_cdc_round1.csv")
feedback_df.to_csv(save_path, index=False)
print("Saved:", save_path.resolve())
```
